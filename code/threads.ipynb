```python
data = pd.read_csv("/Users/lauanton/Coding/convoy/convoy2weeks.csv")
```

```python
conversation_ids = list(data.conversation_id)
```

```python
conversation_ids[:6]
```

```python
queries = []

i = 0
base_query = 'twarc2 search --archive --start-time "2022-01-22T00:00:00" "'
current_query = base_query
while True:
    if i >= len(conversation_ids):
        print("Went thru all conversation ids")
        break
    next_id = conversation_ids[i]
    if len(current_query + "conversation_id:" + str(next_id)) > 1024:
        current_query = current_query[:len(current_query)-5]
        queries.append(current_query + '"') 
        current_query = base_query
    else:
        current_query += "conversation_id:" + str(conversation_ids[i]) + " OR "
        i += 1

print(queries[0:5])
```
